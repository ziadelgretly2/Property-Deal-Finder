```python
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
import pandas as pd

property_type=[]
prices=[]
prop_description=[]
locations=[]
beds_num=[]
baths_num=[]
area=[]
Page_num=1
limit =500
for i in range(limit):
  Page_num+=1

  resullt=requests.get(f'https://www.propertyfinder.eg/en/search?c=1&fu=0&ob=mr&page={Page_num}')
  src=resullt.content
  soup=BeautifulSoup(src,'lxml')

  Type_found=soup.find_all('p',{"class":'styles-module_content__property-type__qxCMa'})
  Price_found=soup.find_all('div',{"class":'styles-module_content__price__TBYWv'})
  Description_found=soup.find_all('h3',{"class":'styles-module_content__title__pLLTh'})
  Location_found=soup.find_all('p',{"class":'styles-module_content__location__yBL3r'})
  Beds_found=soup.find_all('p',{"data-testid":'property-card-spec-bedroom'})
  Bath_found=soup.find_all('p',{"data-testid":'property-card-spec-bathroom'})
  Areas_found=soup.find_all('p',{"data-testid":'property-card-spec-area'})

  min_len = min(len(Type_found), len(Price_found), len(Description_found),
                len(Location_found), len(Beds_found), len(Bath_found), len(Areas_found))

  for j in range(min_len):
    property_type.append(Type_found[j].text)
    prices.append(Price_found[j].text)
    prop_description.append(Description_found[j].text)
    locations.append(Location_found[j].text)
    beds_num.append(Beds_found[j].text)
    baths_num.append(Bath_found[j].text)
    area.append(Areas_found[j].text)

file_list=[property_type,prices,prop_description,locations,beds_num,baths_num,area]
exported=zip_longest(*file_list)

# Fix SyntaxWarning by using a raw string for the file path
with open(r'D:\property.csv','w') as myfile:
  wr=csv.writer(myfile)
  wr.writerow(['property_type','prices','prop_description','locations','beds_num', 'baths_num','area'])
  wr.writerows(exported)
```
